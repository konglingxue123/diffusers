## 使用skimage方法

```python
from skimage.metrics import structural_similarity as ssim
import cv2
import numpy as np
import os
from matplotlib import pyplot as plt
```

```python
def ssim_np_gray(imgs1_dir, imgs2_dir):
    name_list = os.listdir(imgs1_dir)
    ssim_list = []
    for name in name_list:
        img1 = cv2.imread(os.path.join(imgs1_dir, name))
        img2 = cv2.imread(os.path.join(imgs2_dir, name))
        gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
        gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
        ssim_gray, img = ssim(gray1, gray2, full=True, data_range=gray1.max() - gray1.min())  # , full=True用于评估SSIM的全部信息，用图片表示，可视化
        # plt.imshow(img, cmap='gray')
        # print(name, ssim_gray)
        ssim_list.append(ssim_gray)
    return np.mean(ssim_list)
```

```python
def ssim_np_rgb(imgs1_dir, imgs2_dir):
    name_list = os.listdir(imgs1_dir)
    ssim_list = []
    for name in name_list:
        img1 = cv2.imread(os.path.join(imgs1_dir, name))
        img2 = cv2.imread(os.path.join(imgs2_dir, name))
        # print(img1.shape, img2.shape)
        # gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
        # gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
        ssim_rgb, img = ssim(img1, img2, full=True, data_range=img1.max() - img1.min(), channel_axis=2)  # , full=True用于评估SSIM的全部信息，用图片表示，可视化
        # plt.imshow(img, cmap='gray')
        # print(name, ssim_gray)
        ssim_list.append(ssim_rgb)
    return np.mean(ssim_list)
```

```python
# 定义方法，读取文件夹内路径的所有图像，统计为tensor的[N,C,H,W]格式，范围为[0,1]或[0-255]
import os
from PIL import Image
from torchvision import transforms
import torch

def read_images_to_tensor(img_dir, data_range=1.0):
    img_names_list = os.listdir(img_dir)
    img_tensor_list = []
    for img_name in img_names_list:
        img_path = os.path.join(img_dir, img_name)
        if data_range == 1.0:
            img_pil = Image.open(img_path).convert('RGB')
            img_tensor = transforms.ToTensor()(img_pil)
        elif data_range == 255.0:
            img_pil = Image.open(img_path).convert('RGB')
            img_tensor = transforms.ToTensor()(img_pil) * 255.0
        else:
            raise ValueError("data_range should be 1.0 or 255.0")
        img_tensor_list.append(img_tensor)
    img_tensor_batch = torch.stack(img_tensor_list, dim=0)
    return img_tensor_batch
```

```python
def psnr_metric(imgs1_dir, imgs2_dir):
    name_list = os.listdir(imgs1_dir)
    psnr_list = []
    for name in name_list:
        img1 = cv2.imread(os.path.join(imgs1_dir, name))
        img2 = cv2.imread(os.path.join(imgs2_dir, name))

        psnr = compare_psnr(img1, img2)
        # plt.imshow(img, cmap='gray')
        # print(name, ssim_gray)
        psnr_list.append(psnr)
    return np.mean(psnr_list)
```

```python
import os
from PIL import Image
from torchvision import transforms
import torch

def read_images_to_tensor(img_dir, data_range=1.0):
    img_names_list = os.listdir(img_dir)
    img_tensor_list = []
    for img_name in img_names_list:
        img_path = os.path.join(img_dir, img_name)
        if data_range == 1.0:
            img_pil = Image.open(img_path).convert('RGB')
            img_tensor = transforms.ToTensor()(img_pil)
        elif data_range == 255.0:
            img_pil = Image.open(img_path).convert('RGB')
            img_tensor = transforms.ToTensor()(img_pil) * 255.0
        else:
            raise ValueError("data_range should be 1.0 or 255.0")
        img_tensor_list.append(img_tensor)
    img_tensor_batch = torch.stack(img_tensor_list, dim=0)
    return img_tensor_batch
```

## 5. LIPIPS (Learned Perceptual Image Patch Similarity)

```python
from torch import rand
from torchmetrics.image.lpip import LearnedPerceptualImagePatchSimilarity
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
torch.hub.set_dir("/home/ubunto/Project/konglx/generate/diffusers/evaluation_pretrained")

lpips = LearnedPerceptualImagePatchSimilarity(net_type='squeeze', normalize=True)
lpips_alexnet = LearnedPerceptualImagePatchSimilarity(net_type='alex', normalize=True)
lpips_vgg_net = LearnedPerceptualImagePatchSimilarity(net_type='vgg', normalize=True)
# LPIPS needs the images to be in the [-1, 1] range.
# img1 = (rand(50, 3, 512, 512) * 2) - 1
# img2 = (rand(50, 3, 512, 512) * 2) - 1
# lpips(img1, img2)
```

```python
import os
from PIL import Image
from torchvision import transforms
# import torch

def read_images_to_tensor(img_dir, data_range=1.0):
    img_names_list = os.listdir(img_dir)
    img_tensor_list = []
    for img_name in img_names_list:
        img_path = os.path.join(img_dir, img_name)
        if data_range == 1.0:
            img_pil = Image.open(img_path).convert('RGB')
            img_tensor = transforms.ToTensor()(img_pil)
        elif data_range == 255.0:
            img_pil = Image.open(img_path).convert('RGB')
            img_tensor = transforms.ToTensor()(img_pil) * 255.0
        else:
            raise ValueError("data_range should be 1.0 or 255.0")
        img_tensor_list.append(img_tensor)
    img_tensor_batch = torch.stack(img_tensor_list, dim=0)
    return img_tensor_batch
```

```python
def evaluate_each_batch(input_pt1,input_pt2):
    # input_pt1 = input_pt1*2-1
    # input_pt2 = input_pt2*2-1
    
    lpips_list_squeezenet = []
    lpips_list_vgg = []
    lpips_list_alexnet = []
    
    for i in range(input_pt1.shape[0]):
        lpips_squeezenet = lpips(input_pt1[i].unsqueeze(0), input_pt2[i].unsqueeze(0))
        lpips_vgg = lpips_vgg_net(input_pt1[i].unsqueeze(0), input_pt2[i].unsqueeze(0))
        lpips_alex = lpips_alexnet(input_pt1[i].unsqueeze(0), input_pt2[i].unsqueeze(0))
        lpips_list_squeezenet.append(lpips_squeezenet)
        lpips_list_vgg.append(lpips_vgg)    
        lpips_list_alexnet.append(lpips_alex)
    return lpips_list_squeezenet, lpips_list_vgg, lpips_list_alexnet
```
